This notebook contains a custom pytorch model, which will be exported into .ptl format for use in a playtorch application.

```python
# Imports

import torch
import torchvision
from torch.utils.mobile_optimizer import optimize_for_mobile
```

```python
# Mount Google Drive

if 'google.colab' in str(get_ipython()):
  from google.colab import drive
  drive.mount('/content/drive', force_remount=True)

images_path = '/content/drive/My Drive/Colab Notebooks/'
```

```python
# Building (or importing) the model

model = torchvision.models.efficientnet_b0(pretrained=True)
model.eval()
```

```python
# Exporting the model

scripted_model = torch.jit.script(model)
optimized_model = optimize_for_mobile(scripted_model)
optimized_model._save_for_lite_interpreter("/content/drive/My Drive/Colab Notebooks/efficientNet_b0.ptl")

print("model successfully exported")
```
